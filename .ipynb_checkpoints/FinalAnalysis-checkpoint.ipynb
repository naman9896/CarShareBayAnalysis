```python
import pandas as pd
df = pd.read_csv('Car.csv')
df.head()
```

```python
df.tail()
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])
```

```python
# Fill NaN values with 0 for the entire DataFrame
data_filled = df.fillna(0)

# Fill NaN values for numeric columns with the mean
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())

# Fill NaN values for non-numeric columns with a placeholder
non_numeric_columns = df.select_dtypes(exclude=['float64', 'int64']).columns
df[non_numeric_columns] = df[non_numeric_columns].fillna('Unknown')

# Alternatively, drop rows with NaN values if that's preferred
data_cleaned = df.dropna()

```

```python
# Check for duplicate rows
duplicates = df.duplicated()

# Display the number of duplicates
print(f"Number of duplicate rows: {duplicates.sum()}")

# Remove duplicate rows if any
data_cleaned = df.drop_duplicates()
```

```python
# Check the data types of each column
print(df.dtypes)

# Convert data types if necessary, e.g., converting a column to datetime
# Convert 'Date' to datetime, coercing invalid entries to NaT
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Check if the conversion was successful and inspect any NaT values
print(df['Date'].isnull().sum())  # This will show how many 'NaT' values were introduced

# Drop rows with NaT in 'Date' column
df_cleaned = df.dropna(subset=['Date'])
# Fill NaT values with a specific date, like '1900-01-01', or forward fill/backward fill
#df['Date'].fillna(pd.Timestamp('1900-01-01'), inplace=True)  # Replace NaT with a placeholder date
# Fill NaT values with a specific date, such as '1900-01-01', and assign it back to the 'Date' column
df['Date'] = df['Date'].fillna(pd.Timestamp('2019-05-01'))

# Or you can forward fill the missing dates:
# Forward fill missing dates and assign it back to the 'Date' column
#df['Date'] = df['Date'].fillna(method='ffill')

# Check the data types of the 'Date' column
print(df['Date'].dtypes)

# Display a few rows to ensure the conversion is correct
print(df.head())
```

```python
# Save the cleaned data to a new Excel file
data_cleaned.to_csv('cleaned_car.csv', index=False)
```

```python
# Check for any missing or unexpected values in 'Suburb'
print(df['Suburb'].isnull().sum())  # Check for missing values in 'Suburb'
print(df['Suburb'].unique())  # Print unique values in 'Suburb'
# Check for missing values and data types in 'Hours_Booked'
print(df['Hours_Booked'].isnull().sum())  # Check for missing values
print(df['Hours_Booked'].dtype)  # Check data type
# Drop rows where 'Suburb' is NaN
df_cleaned = df.dropna(subset=['Suburb'])
# Convert 'Hours_Booked' to numeric and coerce errors
df_cleaned['Hours_Booked'] = pd.to_numeric(df_cleaned['Hours_Booked'], errors='coerce')
print(df_cleaned.head())
# Drop rows with NaN values in 'Hours_Booked'
df_cleaned = df_cleaned.dropna(subset=['Hours_Booked'])
plt.figure(figsize=(10, 6))
df_cleaned.groupby('Suburb').mean(numeric_only=True)['Hours_Booked'].sort_values(ascending=False).head(10).plot(kind='bar')
plt.title('Average Hours Booked by Suburb')
plt.xlabel('Suburb')
plt.ylabel('Average Hours Booked')
plt.show()
```
